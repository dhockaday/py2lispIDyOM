# Visualizing IDyOM outputs

Given that you already have the `.dat` file output, you can make some figures of outputs for certain melodies (or all melodies).


## 1. Simple plots

`BasicPlot.simple_plot(selected_idyom_output: str,
                      experiment_folder_path: str,
                      melody_names: List[str] = None,
                      starting_index: int = None,
                      ending_index: int = None,
                      savefig: bool = True,
                      showfig: bool = False,
                      fig_format: str = 'png',
                      dpi: float = 400,
                      figsize: tuple = (10, 5))`

This function generates a simple plot with x-axis as the onset value (indicating time),and y-axis as the selected property.

### Example 1.1

In the following example, we will plot the overall (combined) information content of the melody named '"chor-003"' in the experiment logger `16-05-22_14.01.03/`.

We will save the figure and show it here. 

```python
# Import relevant modules
from py2lispIDyOM.viz import BasicPlot

# Set parameters
BasicPlot.simple_plot(selected_idyom_output = 'information.content',
                     experiment_folder_path = 'examples/1_sample_experiment/16-05-22_14.01.03/',
                     melody_names=['"chor-003"'],
                     savefig=True,
                     showfig=True)
```

### Example 1.2

In the following example, we will plot the property 'onset.entropy' of the first 2 melodies in the experiment logger 16-05-22_14.01.03/.

We can also update the default resolution (dpi) and figure size here. And finally we will save the figure and show it here.

```python
# Import relevant modules
from py2lispIDyOM.viz import BasicPlot

# Set parameters
BasicPlot.simple_plot(selected_idyom_output = 'onset.entropy',
                      experiment_folder_path = 'examples/1_sample_experiment/16-05-22_14.01.03/',
                      starting_index = 0,
                      ending_index = 2,
                      savefig=True,
                      showfig=True,
                      dpi = 200,
                      figsize = (10,5))
```

## 2. pitch prediction groundtruth pianoroll plots 


`BasicPlot.pianoroll_pitch_prediction_groundtruth(experiment_folder_path: str,
                                                  melody_names: List[str] = None,
                                                  starting_index: int = None,
                                                  ending_index: int = None,
                                                  savefig: bool = True,
                                                  showfig: bool = False,
                                                  fig_format: str = 'png',
                                                  dpi: float = 400,
                                                  nrows: int = 2,
                                                  ncols: int = 1,
                                                  figsize: tuple = (10, 10))`

This function generates a pair of figures (the predicted pitch distribution and the ground truth) side by side or one on top of another. 

### Example 2.1

In the following example, we will plot the melody named'"chor-007"' in the experiment logger 16-05-22_14.01.03/.

We will save the figure as 'eps' format and show it here.

```python
# Import relevant modules
from py2lispIDyOM.viz import BasicPlot

# Set parameters
BasicPlot.pianoroll_pitch_prediction_groundtruth(experiment_folder_path='examples/1_sample_experiment/16-05-22_14.01.03/',
                                                 melody_names = ['"chor-007"'],
                                                 starting_index = None,
                                                 ending_index = None,
                                                 savefig = True,
                                                 showfig = True,
                                                 fig_format = 'eps')

```

### Example 2.2 

In the following example, we will plot the 8th melody in the experiment logger `16-05-22_14.01.03/`.

We want to show the subplots side-by-side, and will need to ajust the figure size accordingly.

We will save the figure and show it here.

```python
# Import relevant modules
from py2lispIDyOM.viz import BasicPlot

# Set parameters
BasicPlot.pianoroll_pitch_prediction_groundtruth(experiment_folder_path='examples/1_sample_experiment/16-05-22_14.01.03/',
                                                 starting_index = 7,
                                                 ending_index = 8,
                                                 savefig = True,
                                                 showfig = True, 
                                                 nrows = 1, # -> the figure will have one row of subplots
                                                 ncols = 2, # -> the figure will have two column of subplots
                                                 figsize = (20,5) # each subplot is a bit flat, so we adjust the figsize to get better fig quality
                                                )
```

## 3. pianoroll groundtruth and overall surprisal plots


`BasicPlot.pianoroll_groundtruth_overall_surprisal(experiment_folder_path: str,
                                                  melody_names: List[str] = None,
                                                  starting_index: int = None,
                                                  ending_index: int = None,
                                                  savefig: bool = True,
                                                  showfig: bool = False,
                                                  fig_format: str = 'png',
                                                  dpi: float = 400,
                                                  figsize: tuple = (10, 10))`
                                                   
This function plots two subplots: ground truth pianoroll the combined(overall) information content (or surprisal) one on top of another. 


### Example 3.1

In the following example, we will plot the melodies named '"chor-010"' and '"chor-011"' in the experiment logger `16-05-22_14.01.03/`.

We will save the figure and show it here.

```python
# Import relevant modules
from py2lispIDyOM.viz import BasicPlot

# Set parameters
BasicPlot.pianoroll_groundtruth_overall_surprisal(experiment_folder_path='examples/1_sample_experiment/16-05-22_14.01.03/',
                                                 melody_names =['"chor-010"', '"chor-011"'],
                                                 savefig = True,
                                                 showfig = True)

```
