# Visualizing IDyOM outputs

Given that you already have the `.dat` file output, you can make some figures of outputs for certain melodies (or all melodies).


### 1. Simple plots

`BasicPlot.simple_plot(selected_property,
                      experiment_folder_path,
                      melody_names= None,
                      starting_index = None,
                      ending_index = None,
                      savefig = True,
                      showfig = False)`

This function generates a simple plot with x-axis as the onset value (indicating time),and y-axis as the selected property.

In the following example, we will plot the overall (combined) information content (or surprisal) of the melody named '"chor-003"' in the experiment logger `16-05-22_14.01.03/`.

We will save the figure and show it here. 

```python
# Import relevant modules
from modules.viz import BasicPlot

# Set parameters
BasicPlot.simple_plot(selected_property = 'information.content',
                     experiment_folder_path = 'examples/1_sample_experiment/16-05-22_14.01.03/',
                     melody_names=['"chor-003"'],
                     savefig=True,
                     showfig=True)
```

### 2. pitch prediction groundtruth pianoroll plots 


`BasicPlot.pianoroll_pitch_prediction_groundtruth(experiment_folder_path,
                                                 melody_names = None,
                                                 starting_index = None,
                                                 ending_index = None,
                                                 savefig = False,
                                                 showfig = False)`

This function generates a pair of figures (the predicted pitch distribution and the ground truth) side by side. 

In the following example, we will plot the melodies named '"chor-008"' and '"chor-010"' in the experiment logger 16-05-22_14.01.03/.

We will save the figure and show it here.

```python
# Import relevant modules
from modules.viz import BasicPlot

# Set parameters
BasicPlot.pianoroll_pitch_prediction_groundtruth(experiment_folder_path='examples/1_sample_experiment/16-05-22_14.01.03/',
                                                 melody_names = ['"chor-008"', '"chor-010"'],
                                                 starting_index = None,
                                                 ending_index = None,
                                                 savefig = True,
                                                 showfig = True)

```
